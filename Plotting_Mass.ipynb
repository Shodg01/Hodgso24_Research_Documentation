# This notebook is for plotting the mass of the simulations

### It uses txt files that were stored locally in my home directory

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Starting with fiducial data

```python
fid_disk_time, fid_disk_tot_mass, fid_disk_gas_mass, fid_disk_part_mass = np.loadtxt('./txt_files/fid_disk_mass.txt', 
                                                                                     delimiter=',', unpack= True,
                                                                                    skiprows=1)

fid_CGM_time, fid_CGM_tot_mass, fid_CGM_gas_mass, fid_CGM_part_mass, fid_CGM2_tot_mass, fid_CGM2_gas_mass, fid_CGM2_part_mass = np.loadtxt('./txt_files/fid_CGM_mass.txt', 
                                                                                     delimiter=',', unpack= True,
                                                                                    skiprows=1)
```

```python
plt.figure(figsize=(14,5))
plt.suptitle('Disk And CGM Mass Plots Within Virial Radius')

plt.subplot(1,3,1)
plt.title(r'Disk Mass (r$\leq$15 kpc)')
plt.plot(fid_disk_time, fid_disk_tot_mass, color='blue', label='Total Mass')
plt.plot(fid_disk_time, fid_disk_gas_mass, color='orange', ls='--', label='Gas Mass')
plt.plot(fid_disk_time, fid_disk_part_mass, color='red', ls=':', label='Particle Mass')
plt.xlabel('Time [Gyr]')
plt.ylabel(r'Mass [$M_{\odot}$]')
plt.grid(zorder=-10, alpha=0.4)
plt.legend()

plt.subplot(1,3,2)
plt.title(r'CGM Mass (15$<$r$<$206 kpc)')
plt.plot(fid_CGM_time, fid_CGM_tot_mass, color='blue', label='Total Mass')
plt.plot(fid_CGM_time, fid_CGM_gas_mass, color='orange', ls='--', label='Gas Mass')
plt.plot(fid_CGM_time, fid_CGM_part_mass, color='red', ls=':', label='Particle Mass')
plt.xlabel('Time [Gyr]')
plt.ylabel(r'Mass [$M_{\odot}$]')
plt.grid(zorder=-10, alpha=0.4)
plt.legend()

plt.subplot(1,3,3)
plt.title(r'CGM Mass (15$<$r$<$412 kpc)')
plt.plot(fid_CGM_time, fid_CGM2_tot_mass, color='blue', label='Total Mass')
plt.plot(fid_CGM_time, fid_CGM2_gas_mass, color='orange', ls='--', label='Gas Mass')
plt.plot(fid_CGM_time, fid_CGM2_part_mass, color='red', ls=':', label='Particle Mass')
plt.xlabel('Time [Gyr]')
plt.ylabel(r'Mass [$M_{\odot}$]')
plt.grid(zorder=-10, alpha=0.4)
plt.legend()

plt.tight_layout()
```
