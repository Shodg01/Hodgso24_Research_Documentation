# This notebook was a work in progress (WIP) trying to make static in time, projection progressions

### The projections slowly phase through the whole sim, but start out really small.

#### Would be good for an in depth picture of what's happening in the galaxy, starting with almost a slice and then moving to big picture.

The code below is what I've made so far, but it hasn't been tested, so I'm not sure if it actually works.

```python
import yt
import numpy as np
import matplotlib.pyplot as plt
import time

yt.enable_parallelism()

#Load in the dataset you want to work with. Since it's static in time, it will only be for one snapshot,
#unless you put it within the loop for the timeseries
ds = yt.load("/mnt/research/compastro-REU/claire_sims/fiducial/DD0080/DD0080")

#fields you're interested in
fields = ['density', 'temperature', 'metallicity', 'radial_velocity', 'cooling_time']

for field in fields:
#make a while loop to go through different sizes of cylinders to project through

    vr = 206 #kpc   virial radius of sim
    dh = 2
    height = 2
    while height <= vr:
        #loop through x first
        myxcylinder = ds.disk([0.5,0.5,0.5], [1,0,0], (206,'kpc'), (height,'kpc'))
        
        p = yt.ProjectionPlot(ds, 'x', ('gas', field), weight_field=('gas', 'mass'), data_source=myxcylinder,
                              center = [0.5,0.5,0.5])
        p.zoom(3)
        p.annotate_timestamp(text_args={'color':'dimgray'})
        
        #need to set colorbar to stay constant. May need to add or change if have more/different fields
        if field == 'density':
            p.set_zlim(("gas", field), zmin=(1e-30, "g/cm**3"), zmax=(1e-24, "g/cm**3"))
        elif field == 'temperature':
            p.set_zlim(("gas", field), zmin=(1e3, "K"), zmax=(1e8, "K"))
        elif field == 'metallicity':
            p.set_zlim(("gas", field), zmin=(1e-1, "Zsun"), zmax=(1e1, "Zsun"))
        elif field == 'radial_velocity':
            p.set_zlim(("gas", field), zmin=(1e-7, "cm/s"), zmax=(1e8, "cm/s"))
        elif field == 'cooling_time':
            p.set_zlim(("gas", field), zmin=(1e15, "s"), zmax=(1e24, "s"))
        
        #save the projection
        p.save("./Fiducial_"+str(field)+"_projection_x_progression_height"+str(height)+".png")
        
        ######################################
        #loop through z now
        
        myzcylinder = ds.disk([0.5,0.5,0.5], [0,0,1], (206,'kpc'), (height,'kpc'))
        
        p = yt.ProjectionPlot(ds, 'z', ('gas', field), weight_field=('gas', 'mass'), data_source=myzcylinder,
                              center = [0.5,0.5,0.5])
        p.zoom(3)
        p.annotate_timestamp(text_args={'color':'dimgray'})
        
        #need to set colorbar to stay constant. May need to add or change if have more/different fields
        if field == 'density':
            p.set_zlim(("gas", field), zmin=(1e-30, "g/cm**3"), zmax=(1e-24, "g/cm**3"))
        elif field == 'temperature':
            p.set_zlim(("gas", field), zmin=(1e3, "K"), zmax=(1e8, "K"))
        elif field == 'metallicity':
            p.set_zlim(("gas", field), zmin=(1e-1, "Zsun"), zmax=(1e1, "Zsun"))
        elif field == 'radial_velocity':
            p.set_zlim(("gas", field), zmin=(1e-7, "cm/s"), zmax=(1e8, "cm/s"))
        elif field == 'cooling_time':
            p.set_zlim(("gas", field), zmin=(1e15, "s"), zmax=(1e24, "s"))
        
        #save the projection
        p.save("./Fiducial_"+str(field)+"_projection_z_progression_height"+str(height)+".png")
        
        height += dh
```
