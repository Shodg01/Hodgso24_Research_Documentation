```python
###############################################
# Script for making plots of cumulative 
# star formation. Plots the cumulative star 
# plot at the end of the code, could change 
# if you want just the data
###############################################

# imports
import yt
import matplotlib.pyplot as plt
from yt.data_objects.particle_filters import add_particle_filter
import numpy as np

# make the timeseries
fid_ts = yt.load("/mnt/research/galaxies-REU/sims/claire_isogal_sims/fiducial/DD????/DD????")

# make the list of total stars formed
fiducial_total_stars_formed = []
fiducial_times_count = []

# loop through all the snapshots in the sim
for ds in fid_ts.piter():
    
    # making sure there are particles to run it
    # if there aren't set the total stars to previous value or zero if start
    if ds.particles_exist == False:
        continue
        
    else:
        if len(fiducial_total_stars_formed) == 0:
            fiducial_total_stars_formed.append(0)
            current_time = ds.current_time.to('Gyr')
            fiducial_times_count.append(current_time)
        else:
            fiducial_total_stars_formed.append(fiducial_total_stars_formed[-1])
            current_time = ds.current_time.to('Gyr')
            fiducial_times_count.append(current_time)
        
    current_time = ds.current_time.to('Gyr')
    fiducial_times_count.append(current_time)
    
    def formed_star(pfilter, data):
        filter = data['all', 'creation_time'] >0
        return filter

    add_particle_filter('formed_star', function=formed_star, filtered_type='all', requires=['creation_time'])

    ds.add_particle_filter("formed_star")
    ad = ds.all_data()
    masses = ad["formed_star", "particle_mass"].in_units("Msun")    
    
    total_mass = sum(masses)
    fiducial_total_stars_formed.append(total_mass)
    
    
################################################

# make the timeseries
h2sf_ts = yt.load("/mnt/research/galaxies-REU/sims/claire_isogal_sims/h2sf/DD????/DD????")

# make the list of total stars formed
h2sf_total_stars_formed = []
h2sf_times_count = []

# loop through all the snapshots in the sim
for ds in h2sf_ts.piter():
    
    # making sure there are particles to run it
    # if there aren't set the total stars to previous value or zero if start
    if ds.particles_exist == False:
        continue
    else:
        if len(h2sf_total_stars_formed) == 0:
            h2sf_total_stars_formed.append(0)
            current_time = ds.current_time.to('Gyr')
            h2sf_times_count.append(current_time)
        else:
            h2sf_total_stars_formed.append(h2sf_total_stars_formed[-1])
            current_time = ds.current_time.to('Gyr')
            h2sf_times_count.append(current_time)
        
    current_time = ds.current_time.to('Gyr')
    h2sf_times_count.append(current_time)
    
    def formed_star(pfilter, data):
        filter = data['all', 'creation_time'] >0
        return filter

    add_particle_filter('formed_star', function=formed_star, filtered_type='all', requires=['creation_time'])

    ds.add_particle_filter("formed_star")
    ad = ds.all_data()
    masses = ad["formed_star", "particle_mass"].in_units("Msun")    
    
    total_mass = sum(masses)
    h2sf_total_stars_formed.append(total_mass)  
    
######################################################    
    
# make the timeseries
h2sf_lowFB_ts = yt.load("/mnt/research/galaxies-REU/sims/claire_isogal_sims/h2sf_lowFB/DD????/DD????")

# make the list of total stars formed
h2sf_lowFB_total_stars_formed = []
h2sf_lowFB_times_count = []

# loop through all the snapshots in the sim
for ds in h2sf_lowFB_ts.piter():
    
    # making sure there are particles to run it
    # if there aren't set the total stars to previous value or zero if start
    if ds.particles_exist == False:
        continue
    else:
        if len(h2sf_lowFB_total_stars_formed) == 0:
            h2sf_lowFB_total_stars_formed.append(0)
            current_time = ds.current_time.to('Gyr')
            h2sf_lowFB_times_count.append(current_time)
        else:
            h2sf_lowFB_total_stars_formed.append(h2sf_lowFB_total_stars_formed[-1])
            current_time = ds.current_time.to('Gyr')
            h2sf_lowFB_times_count.append(current_time)
        
    filename = ds.filename.split('/')[-1]
    current_time = ds.current_time.to('Gyr')
    h2sf_lowFB_times_count.append(current_time)
    
    def formed_star(pfilter, data):
        filter = data['all', 'creation_time'] >0
        return filter

    add_particle_filter('formed_star', function=formed_star, filtered_type='all', requires=['creation_time'])

    ds.add_particle_filter("formed_star")
    ad = ds.all_data()
    masses = ad["formed_star", "particle_mass"].in_units("Msun")    
    
    total_mass = sum(masses)
    h2sf_lowFB_total_stars_formed.append(total_mass)
    
############################################################

# make the timeseries
newCGM_ts = yt.load("/mnt/research/galaxies-REU/sims/claire_isogal_sims/newCGM_h2sf/DD????/DD????")

# make the list of total stars formed
newCGM_total_stars_formed = []
newCGM_times_count = []

# loop through all the snapshots in the sim
for ds in newCGM_ts.piter():
    
    # making sure there are particles to run it
    # if there aren't set the total stars to previous value or zero if start
    if ds.particles_exist == False:
        continue
    else:
        if len(newCGM_total_stars_formed) == 0:
            newCGM_total_stars_formed.append(0)
            current_time = ds.current_time.to('Gyr')
            newCGM_times_count.append(current_time)
        else:
            newCGM_total_stars_formed.append(newCGM_total_stars_formed[-1])
            current_time = ds.current_time.to('Gyr')
            newCGM_times_count.append(current_time)
        
    filename = ds.filename.split('/')[-1]
    current_time = ds.current_time.to('Gyr')
    newCGM_times_count.append(current_time)
    
    def formed_star(pfilter, data):
        filter = data['all', 'creation_time'] >0
        return filter

    add_particle_filter('formed_star', function=formed_star, filtered_type='all', requires=['creation_time'])

    ds.add_particle_filter("formed_star")
    ad = ds.all_data()
    masses = ad["formed_star", "particle_mass"].in_units("Msun")    
    
    total_mass = sum(masses)
    newCGM_total_stars_formed.append(total_mass)
    

####################################################

# change the lists to arrays
fiducial_cumulative_star_count = np.array(fiducial_total_stars_formed)
fiducial_times = np.array(fiducial_times_count)

h2sf_cumulative_star_count = np.array(h2sf_total_stars_formed)
h2sf_times = np.array(h2sf_times_count)

h2sf_lowFB_cumulative_star_count = np.array(h2sf_lowFB_total_stars_formed)
h2sf_lowFB_times = np.array(h2sf_lowFB_times_count)

newCGM_cumulative_star_count = np.array(newCGM_total_stars_formed)
newCGM_times = np.array(newCGM_times_count)

# make the txt files
combined_fid_data = np.column_stack((fiducial_times,fiducial_cumulative_star_count))
np.savetxt('Fid_Stars_Formed_data.txt', combined_fid_data, delimiter=',', 
            header='fiducial_times,fiducial_cumulative_star_count', 
           comments='', fmt='%.6e')
combined_h2sf_data = np.column_stack((h2sf_times,h2sf_cumulative_star_count))
np.savetxt('h2sf_Stars_Formed_data.txt', combined_h2sf_data, delimiter=',', 
            header='h2sf_times,h2sf_cumulative_star_count', 
           comments='', fmt='%.6e')
combined_lowFB_data = np.column_stack((h2sf_lowFB_times,h2sf_lowFB_cumulative_star_count))
np.savetxt('lowFB_Stars_Formed_data.txt', combined_lowFB_data, delimiter=',', 
            header='lowFB_times,lowFB_cumulative_star_count', 
           comments='', fmt='%.6e')
combined_newCGM_data = np.column_stack((newCGM_times,newCGM_cumulative_star_count))
np.savetxt('newCGM_Stars_Formed_data.txt', combined_newCGM_data, delimiter=',', 
            header='newCGM_times,newCGM_cumulative_star_count', 
           comments='', fmt='%.6e')

# plot the data and save it
plt.figure(figsize=(15,10))
plt.plot(fiducial_times, fiducial_cumulative_star_count, label='Fiducial')
plt.plot(h2sf_times, h2sf_cumulative_star_count, label='h2sf')
plt.plot(h2sf_lowFB_times, h2sf_lowFB_cumulative_star_count, label='h2sf_lowFB')
plt.plot(newCGM_times, newCGM_cumulative_star_count, label='newCGM')
plt.xlabel('Time (Gyr)')
plt.ylabel('Stellar Mass Formed (Msun)')
plt.grid()
plt.legend()
plt.savefig('/mnt/scratch/hodgso24/plots/Cumulative_Star_Plot_Redo.png')
```
