# This notebook is for plotting the mean mass weighted metallicity and temperature as a function of time for the h2sf and the newCGM_h2sf simulations

```python
##############################
# Script for plotting 
# mean mass weighted 
# metallicity and temperature
##############################

# imports
import yt
import numpy as np
import matplotlib.pyplot as plt

# make the timeseries
ts_h2sf = yt.load("/mnt/research/galaxies-REU/sims/claire_isogal_sims/h2sf/DD????/DD????")
ts_newCGM = yt.load("/mnt/research/galaxies-REU/sims/claire_isogal_sims/newCGM_h2sf/DD????/DD????")

# make the master lists
h2sf_disk_metallicity = []
newCGM_disk_metallicity = []
h2sf_CGM_metallicity = []
newCGM_CGM_metallicity = []
h2sf_CGM2_metallicity = []
newCGM_CGM2_metallicity = []
h2sf_ad_metallicity = []
newCGM_ad_metallicity = []

h2sf_disk_temp = []
newCGM_disk_temp = []
h2sf_CGM_temp = []
newCGM_CGM_temp = []
h2sf_CGM2_temp = []
newCGM_CGM2_temp = []
h2sf_ad_temp = []
newCGM_ad_temp = []

h2sf_time = []
newCGM_time = []

field_1 = ("gas", "metallicity")  # The field to average
field_2 = ("gas", "temperature")  # The field to average
weight = ("gas", "mass")  # The weight for the average

# main for loop for h2sf sim
for ds in ts_h2sf.piter():
    # make objects for disk and various CGM
    disk = ds.disk([0.5,0.5,0.5], [0,0,1], (25, 'kpc'), (5, 'kpc'))
    CGM = ds.sphere([0.5,0.5,0.5], (206, "kpc"))
    CGM2 = ds.sphere([0.5,0.5,0.5], (412, "kpc"))
    ad = ds.all_data()
    
    # get weighted averages
    average_disk_metallicity = disk.quantities.weighted_average_quantity(field_1, weight).in_units('Zsun')
    average_disk_temperature = disk.quantities.weighted_average_quantity(field_2, weight)
    
    average_CGM_metallicity = CGM.quantities.weighted_average_quantity(field_1, weight).in_units('Zsun')
    average_CGM_temperature = CGM.quantities.weighted_average_quantity(field_2, weight)
    
    average_CGM2_metallicity = CGM2.quantities.weighted_average_quantity(field_1, weight).in_units('Zsun')
    average_CGM2_temperature = CGM2.quantities.weighted_average_quantity(field_2, weight)
    
    average_ad_metallicity = ad.quantities.weighted_average_quantity(field_1, weight).in_units('Zsun')
    average_ad_temperature = ad.quantities.weighted_average_quantity(field_2, weight)
    
    # add to lists
    h2sf_disk_metallicity.append(average_disk_metallicity)
    h2sf_disk_temp.append(average_disk_temperature)
    
    h2sf_CGM_metallicity.append(average_CGM_metallicity)
    h2sf_CGM_temp.append(average_CGM_temperature)
    
    h2sf_CGM2_metallicity.append(average_CGM2_metallicity)
    h2sf_CGM2_temp.append(average_CGM2_temperature)
    
    h2sf_ad_metallicity.append(average_ad_metallicity)
    h2sf_ad_temp.append(average_ad_temperature)
    
    current_time = ds.current_time.to('Gyr')
    h2sf_time.append(current_time)
    

# main for loop for newCGM sim
for ds in ts_newCGM.piter():
    # make objects for disk and various CGM
    disk = ds.disk([0.5,0.5,0.5], [0,0,1], (22, 'kpc'), (5, 'kpc'))
    CGM = ds.sphere([0.5,0.5,0.5], (206, "kpc"))
    CGM2 = ds.sphere([0.5,0.5,0.5], (412, "kpc"))
    ad = ds.all_data()
    
    # get weighted averages
    average_disk_metallicity = disk.quantities.weighted_average_quantity(field_1, weight).in_units('Zsun')
    average_disk_temperature = disk.quantities.weighted_average_quantity(field_2, weight)
    
    average_CGM_metallicity = CGM.quantities.weighted_average_quantity(field_1, weight).in_units('Zsun')
    average_CGM_temperature = CGM.quantities.weighted_average_quantity(field_2, weight)
    
    average_CGM2_metallicity = CGM2.quantities.weighted_average_quantity(field_1, weight).in_units('Zsun')
    average_CGM2_temperature = CGM2.quantities.weighted_average_quantity(field_2, weight)
    
    average_ad_metallicity = ad.quantities.weighted_average_quantity(field_1, weight).in_units('Zsun')
    average_ad_temperature = ad.quantities.weighted_average_quantity(field_2, weight)
    
    # add to lists
    newCGM_disk_metallicity.append(average_disk_metallicity)
    newCGM_disk_temp.append(average_disk_temperature)
    
    newCGM_CGM_metallicity.append(average_CGM_metallicity)
    newCGM_CGM_temp.append(average_CGM_temperature)
    
    newCGM_CGM2_metallicity.append(average_CGM2_metallicity)
    newCGM_CGM2_temp.append(average_CGM2_temperature)
    
    newCGM_ad_metallicity.append(average_ad_metallicity)
    newCGM_ad_temp.append(average_ad_temperature)
    
    current_time = ds.current_time.to('Gyr')
    newCGM_time.append(current_time)
    
# change to arrays    
h2sf_disk_metal_array = np.array(h2sf_disk_metallicity)
newCGM_disk_metal_array = np.array(newCGM_disk_metallicity)

h2sf_CGM_metal_array = np.array(h2sf_CGM_metallicity)
newCGM_CGM_metal_array = np.array(newCGM_CGM_metallicity)

h2sf_CGM2_metal_array = np.array(h2sf_CGM2_metallicity)
newCGM_CGM2_metal_array = np.array(newCGM_CGM2_metallicity)

h2sf_ad_metal_array = np.array(h2sf_ad_metallicity)
newCGM_ad_metal_array = np.array(newCGM_ad_metallicity)

h2sf_disk_temp_array = np.array(h2sf_disk_temp)
newCGM_disk_temp_array = np.array(newCGM_disk_temp)

h2sf_CGM_temp_array = np.array(h2sf_CGM_temp)
newCGM_CGM_temp_array = np.array(newCGM_CGM_temp)

h2sf_CGM2_temp_array = np.array(h2sf_CGM2_temp)
newCGM_CGM2_temp_array = np.array(newCGM_CGM2_temp)

h2sf_ad_temp_array = np.array(h2sf_ad_temp)
newCGM_ad_temp_array = np.array(newCGM_ad_temp)

h2sf_time_array = np.array(h2sf_time)
newCGM_time_array = np.array(newCGM_time)

# combine the data and save as txt
h2sf_combined_data = np.column_stack((h2sf_time_array, h2sf_disk_metal_array, h2sf_CGM_metal_array, h2sf_CGM2_metal_array, h2sf_ad_metal_array, h2sf_disk_temp_array, h2sf_CGM_temp_array, h2sf_CGM2_temp_array, h2sf_ad_temp_array))
np.savetxt('h2sf_weighted_averages.txt', h2sf_combined_data, delimiter=',',
           header='time,disk metal,CGM metal,CGM2 metal,ad metal,disk temp,CGM temp,CGM2 temp,ad temp',
           comments='', fmt='%.6e')

newCGM_combined_data = np.column_stack((newCGM_time_array, newCGM_disk_metal_array, newCGM_CGM_metal_array, newCGM_CGM2_metal_array, newCGM_ad_metal_array, newCGM_disk_temp_array, newCGM_CGM_temp_array, newCGM_CGM2_temp_array, newCGM_ad_temp_array))
np.savetxt('newCGM_weighted_averages.txt', newCGM_combined_data, delimiter=',',
           header='time,disk metal,CGM metal,CGM2 metal,ad metal,disk temp,CGM temp,CGM2 temp,ad temp',
           comments='', fmt='%.6e')
```
